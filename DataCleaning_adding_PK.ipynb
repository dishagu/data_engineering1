### For Batting Table

```python
import pandas as pd
```

```python
file_path = 'C:/Users/Manoj/Downloads/final_csv/BattingSummaryFinal.csv'  #import the priviously cleaned file
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
# Adding Inning_num column
df['Inning_num'] = df['Innings'].apply(lambda x: 1 if x == 'First Innings' else 2)

# Adding batting position column
df['Batting_pos'] = df.groupby(['MatchID', 'Inning_num']).cumcount() + 1

# Creating primary key
df['id'] = df['MatchID'].astype(str) + df['Inning_num'].astype(str) +  df['Batting_pos'].astype(str)
```

### For Bowler Table

```python
bowler_path = 'C:/Users/Manoj/Downloads/final_csv/BowlingSummary.csv'  # import priviously cleaned bowler table
df2 = pd.read_csv(bowler_path)
```

```python
df2.head()
```

```python
# Adding Inning_num column
df2['Inning_num'] = df2['Innings_Bowled'].apply(lambda x: 1 if x == 'First Innings' else 2)

# Adding bowling position column
df2['Bowling_pos'] = df2.groupby(['Match_ID', 'Inning_num']).cumcount() + 1

# Creating primary key
df2['id'] = df2['Match_ID'].astype(str) + df2['Inning_num'].astype(str) + df2['Bowling_pos'].astype(str)
```
