```python
# Directory to save images
image_dir = 'players'
os.makedirs(image_dir, exist_ok=True)
```

```python
# Initialize the WebDriver
driver = webdriver.Chrome(executable_path=webdriver_path)
```

```python
# Base URL for Cricbuzz search
base_url = 'https://www.cricbuzz.com/search'
```

```python
def get_player_image_url(player_name):
    try:
        # Navigate to the Cricbuzz search page
        driver.get(base_url)
        
        # Find the search input box and enter the player's name
        search_box = driver.find_element(By.ID, 'search_bar')
        search_box.clear()
        search_box.send_keys(player_name)
        
        # Find and click the search button
        search_button = driver.find_element(By.XPATH, '//*[@id="search_bar"]')
        search_button.click()
        
        # Wait for the search results to load
        time.sleep(7)  # Adjust this as necessary for your internet speed
        
        # Parse the search results page
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        # Find the player card which should contain the image
        img_tag = soup.find('img', class_='cb-srch-suggest-icon-img')
        
        if img_tag and 'src' in img_tag.attrs:
            img_url = img_tag['src']
            
            # Download the image using requests
            img_response = requests.get(img_url)
            
            # Check if the request was successful
            if img_response.status_code == 200:
                # Save the image to a file
                with open(f"{player_name.replace(' ', '_')}.jpg", 'wb') as f:
                    f.write(img_response.content)
                
                return img_url
            else:
                print(f"Failed to download image. Status code: {img_response.status_code}")
                
        
    
    except Exception as e:
        print(f"Exception occurred: {str(e)}")
    
    return None
```

```python
def download_image(img_url, player_name):
    response = requests.get(img_url)
    file_path = os.path.join(image_dir, f"{player_name.replace(' ', '_')}.jpg")
    with open(file_path, 'wb') as f:
        f.write(response.content)
        
```
